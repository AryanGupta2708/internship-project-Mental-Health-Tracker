This notebook deals with the 'Mental Fitness' using Machine Learning Algortithm: Random Forest Regression

## Import Libraries

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np #Linear Algebra
import pandas as pd #Date Processing
import seaborn as sns #Data Visualization
import matplotlib.pyplot as plt #Visualization Utility: Graph Plot
from sklearn.preprocessing import LabelEncoder #Normalize Labels
from sklearn.model_selection import train_test_split #Split Data Into Training And Testing Data
from sklearn.ensemble import RandomForestRegressor #Machine Learning Algorithm
```

## Import Data 

```python
df1= pd.read_csv("MentalHealthData01.csv")
df1.head(11)
```

```python
df2= pd.read_csv("MentalHealthData02.csv")
df2.head(11)
```

## Merge Data 

```python
data= pd.merge(df1, df2)
data.head(11)
```

## Data Cleaning 

```python
data.isnull().sum()
```

```python
data.drop('Code', axis=1, inplace=True)
```

## Data Visualization 

```python
data.set_axis(['Country', 'Year', 
               'Schizophrenia', 'BipolarDisorder', 'EatingDisorder', 'Anxiety', 
               'DrugUsage', 'Depression', 'Alcohol', 
               'MentalFitness'], axis='columns', inplace= True)
data.head(11)
```

Pairplot is a data visualization technique that displays pairwise relationships between multiple variables in a dataset, typically using scatter plots and histograms, providing a comprehensive overview of the data's distributions and correlations.

```python
plt.figure(figsize=(15,12))
sns.pairplot(data[['Country', 'Year', 
               'Schizophrenia', 'BipolarDisorder', 'EatingDisorder', 'Anxiety', 
               'DrugUsage', 'Depression', 'Alcohol', 
               'MentalFitness']])
```

## Transform Data 

```python
l= LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'object':
        data[i]= l.fit_transform(data[i])
```

# Machine Learning Algorithm : Random Forest Regression

RandomForestRegression is a machine learning algorithm that combines multiple decision trees to make predictions by averaging the outputs of individual trees, providing a robust and accurate method for regression tasks.

## Split Data

```python
x= data.drop('MentalFitness', axis=1)
y= data['MentalFitness']

xtrain, xtest, ytrain, ytest= train_test_split(x, y)
```

## Model Implementation

```python
ml=RandomForestRegressor()
ml.fit(xtrain, ytrain)
```

## Training Data

```python
ytrain_predict=ml.predict(xtrain)
```

```python
plt.figure(figsize=(15, 12))
plt.plot(range(100), ytrain[:100], label='True')
plt.plot(range(100), ytrain_predict[:100], label='Predicted')
plt.legend()
plt.title('True vs Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

## Testing Data

```python
ytest_predict= ml.predict(xtest)
```

```python
plt.figure(figsize=(15, 12))
plt.plot(range(100), ytest[:100], label='True')
plt.plot(range(100), ytest_predict[:100], label='Predicted')
plt.legend()
plt.title('True vs Predicted Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```
